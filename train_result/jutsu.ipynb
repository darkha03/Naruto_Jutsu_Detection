```python
import shutil
import os
from ultralytics import YOLO

# 1. COPY DATA TO THE WRITABLE FOLDER
input_path = '/kaggle/input/datasets/karlmorg37/jutsu-v6'
working_path = '/kaggle/working/Jutsu-6'

if not os.path.exists(working_path):
    print("Copying data to writable folder... This might take a minute.")
    shutil.copytree(input_path, working_path)
    print("Copy complete!")

# 2. LOAD THE BRAIN
model = YOLO('yolov8m.pt')

# 3. TRAIN! 
results = model.train(
    data=f'{working_path}/data.yaml',
    epochs=80,
    imgsz=640,
    augment=True,
    label_smoothing=0.0,
    cls=1.5,
)
```

```python
# 1. Load trained Brain
model = YOLO('/kaggle/working/runs/detect/train/weights/best.pt')

# 2. Run the final exam strictly on the 'test' dataset
print("Starting the final exam...")
metrics = model.val(data=f'{working_path}/data.yaml', split='test')

# 3. Print the final grades
print("--- FINAL EXAM RESULTS ---")
print(f"Overall Accuracy (mAP50): {metrics.box.map50 * 100:.2f}%")
print(f"Precision: {metrics.box.mp * 100:.2f}%")
print(f"Recall: {metrics.box.mr * 100:.2f}%")
```
